```python
%matplotlib inline
import cv2
from matplotlib import pyplot as plt
import numpy as np
import time as t

print ("OpenCV Version : " + cv2.__version__)
```

```python
PATH = "kirito-and-asuna-wallpaper-2.jpg"
image = cv2.imread(PATH)
```

```python
##using matplotlib
plt.imshow(image)
plt.xticks([]), plt.yticks([])  # to hide tick values on X and Y axis
plt.show()
```

```python
## revert BGR to RDB to see original image:
cv2_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(cv2_rgb)
plt.show()
```

```python
## display using cv2:
cv2_ = cv2.imread(PATH, 1) ##1 - normal, 0 - grayscale, -1 - Loads image as such including alpha channel
cv2.imshow('image', cv2_)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
## gray using matplotlib
img = cv2.imread(PATH, 0)
plt.imshow(img, cmap = 'gray', interpolation = 'bicubic')
plt.xticks([]), plt.yticks([])  ## to hide tick values on X and Y axis
plt.show()
```
