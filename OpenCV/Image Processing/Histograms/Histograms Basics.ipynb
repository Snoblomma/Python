```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("stil_life.jpg", 0)
hist = cv2.calcHist([img],[0],None,[256],[0,256])
```

### Calculating Histogram in cv2

```python
##[0] - channels, [0] for grey, [1] for colour, [2] for BRG
##None - mask image
##[256] - size of histogram, 256 max
##[0,256] - range
hist = cv2.calcHist([img],[0],None,[256],[0,256])
```

### Calculating Histogram using numpy
Here histogram is a 256x1 array, each value corresponds to number of pixels in that image with its corresponding pixel value.

```python
hist,bins = np.histogram(img.ravel(), 256, [0, 256])
```

### Plotting histogram using pyplot

```python
img = cv2.imread('church.jpg',0)

plt.hist(img.ravel(),256,[0,256]); 
plt.show()
```

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread('church.jpg')
color = ('b','g','r')
for i,col in enumerate(color):
    histr = cv2.calcHist([img],[i],None,[256],[0,256])
    plt.plot(histr,color = col)
    plt.xlim([0,256])
plt.show()
```
