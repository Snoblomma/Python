```python
import cv2
import numpy as np
from matplotlib import pyplot as plt

img = cv2.imread("vase.jpg", 0)
imgL = cv2.imread("vase.jpg", 0)
imgR = cv2.imread("vase.jpg", 0)

stereo = cv2.StereoBM_create(numDisparities=16, blockSize=5)
disparity = stereo.compute(imgL,imgR)

plt.subplot(121)
plt.imshow(img)
plt.title('Original')
plt.xticks([]), plt.yticks([])

plt.subplot(122)
plt.imshow(disparity)
plt.title('Depth map')
plt.xticks([]), plt.yticks([])

plt.show()
```
