```python
import os
import urllib.request
from urllib.parse import urljoin
from bs4 import BeautifulSoup as BS

base_url = "http://www.gutenberg.org/files/10625/10625-h/dict2.html"

content = urllib.request.urlopen(base_url).read()

soup = BS(content, "lxml").findAll("b", class_= "entry")
    
print(len(soup))

r = list()

for s in soup:
    d = s.getText()
    if d not in r:
        r.append(d)
    
for item in r:
    print(item)
```
