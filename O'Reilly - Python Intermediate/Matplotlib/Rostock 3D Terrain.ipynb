### Rostock City Terrain

```python
from mpl_toolkits.mplot3d.axes3d import *
from matplotlib import cm
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.mlab import griddata

x = []
y = []
z = []

data = open("rostock.csv", "r").readlines()

for line in data:
    d = line.strip().split(",")
    x.append(float(d[0]))
    y.append(float(d[1]))
    z.append(float(d[2]))

fig = plt.figure()
ax = Axes3D(fig)
ax.scatter3D(x,y,z, c = z, cmap=plt.cm.jet)  
ax.view_init(30, 120) ## better view


##setting size:
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.show()
```

```python
import numpy as np
from matplotlib.mlab import griddata
# craation of a 2D grid
xi = np.linspace(min(x), max(x))
yi = np.linspace(min(y), max(y))
X, Y = np.meshgrid(xi, yi)
# interpolation
Z = griddata(x, y, z, xi, yi, interp='linear')
fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,linewidth=1, antialiased=True)
ax.view_init(30, 120)
plt.show()
```

```python

ax = plt.subplot()
ax.tricontourf(x,y,z, 40) # choose 20 contour levels, just to show how good its interpolation is
ax.plot(x,y, 'o')

plt.show()
```

```python
import scipy as sp
import scipy.interpolate
# 2D grid construction
spline = sp.interpolate.Rbf(x,y,z,function='thin-plate')
xi = np.linspace(min(x), max(x))
yi = np.linspace(min(y), max(y))
X, Y = np.meshgrid(xi, yi)
# interpolation
Z = spline(X,Y)

fig = plt.figure()
ax = Axes3D(fig)
ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=cm.jet,linewidth=1, antialiased=True)
ax.view_init(30, 120)
plt.show()
```
