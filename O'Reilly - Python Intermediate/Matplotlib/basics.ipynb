### Displaying World Polpulaion

```python
data = open("world_population.txt", "r").readlines()
dates = []
populations = []

for point in data:
    d = point.split()
    date = d[0]
    population = d[1].replace(",", "")
    dates.append(date)
    populations.append(population)
    
pyplot.plot(dates, populations, "-")
pyplot.ylabel("World population")
pyplot.xlabel("Year")
pyplot.title("World population over time")
pyplot.show()
```

### Life expectancy

```python
from matplotlib import pyplot as plt

data = open("life_expectancy.txt", "r").readlines()
women = []
men = []
years = []

for point in data:
    d = point.split()
    years.append(d[0])
    men.append(d[2])
    women.append(d[3])
    
plt.plot(years, women, "-", label = "Life expectancy of women")
plt.plot(years, men, "-", label = "Life expectancy of men")
plt.legend(loc='best')
plt.ylabel("Life expectancy")
plt.xlabel("Year")
plt.title("Life expactancy in US by years")

##setting size:
# Get current size
fig_size = plt.rcParams["figure.figsize"]
 
# Set figure width to 12 and height to 9
fig_size[0] = 12
fig_size[1] = 9
plt.rcParams["figure.figsize"] = fig_size

plt.show()
```

### Deaths by Major Causes, 1960-2011

```python
data = open("death_by_cause.txt", "r").readlines()
key = open("death_by_cause_key.txt", "r").readlines()

for i in range(0, len(key)):
    years = []
    current_key = []
    for d in data:
        sp = d.split()
        years.append(sp[0])
        current_key.append(sp[i+1])
    plt.plot(years, current_key, "-", label = key[i])
  

plt.legend(loc='best')
plt.ylabel("Cause")
plt.xlabel("Year")
plt.title("Deaths by Major Causes, 1960-2011")
plt.show()
```

### Pie Charts

```python
# Data to plot
labels = 'Python', 'C++', 'Ruby', 'Java'
sizes = [215, 130, 245, 210]
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
explode = (0.2, 0, 0, 0)  # explode 1st slice
 
# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=140)
 
plt.axis('equal')
plt.show()
```
