```python
f = open("sonnets.txt", "r")
s = open("words.txt", "w")

words = []

for line in f:
    entries = line.strip('\n\t"0123456789,.;:?!').split(" ")
    for entry in entries:
            entry = entry.strip('\n\t"0123456789,.;:?!').upper()
            if entry not in words:
                words.append(entry)

words.sort()
words

for word in words:
    s.write(word)
    
f.close()
s.close()
```
