### Displaying World Polpulaion

```python
data = open("world_population.txt", "r").readlines()
dates = []
populations = []

for point in data:
    d = point.split()
    date = d[0]
    population = d[1].replace(",", "")
    dates.append(date)
    populations.append(population)
    
pyplot.plot(dates, populations, "o-")
pyplot.ylabel("World population")
pyplot.xlabel("Year")
pyplot.title("World population over time")
pyplot.show()
```
