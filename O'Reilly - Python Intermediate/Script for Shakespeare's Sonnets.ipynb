```python
import string

sonnets = open("sonnets.txt").readlines()
word_set = set([elt.strip for elt in open("scrabble.txt")])
sonnet_words = set()

def strip_punctuation(word):
    ## remove punctuation
    ## if there id apostrophe in the word, skip it
    word.replace("-", " ")
    apostrophe_index = word.find("'")
    if apostrophe_index != -1:
        return None
    return word.strip(string.punctuation)

for line in sonnets:
    ## split apart hyphenated words
    line_words = line.replace("-", " ").strip().split()
    
    ## if its an empty line or sonnet number, skip it
    if len(line_words) <= 1:
        continue
        
    for word in line_words:
        stripped = strip_punctuation(word)
        if stripped and len(stripped) > 1:
            sonnet_words.add(stripped.upper())
            
sonnet_words = list(sonnet_words)
sonnet_words.sort()

for w in sonnet_words:
    print(w)
```
